# K-means Model dividing Mall customers into clusters 

The aim of this notebook is to divide customers of a shopping mall into clusters to make it easier to target marketing campaigns to specific groups of shoppers

We are going to be using the data on customers provided by the mall that will include their age, spending score and Annual income

```python
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
customers = pd.read_csv('Mall_Customers.csv')
```

```python
customers
```

```python
model = KMeans(n_clusters=5)
```

```python
cols_to_keep = ['Annual Income (k$)', 'Spending Score (1-100)']
```

```python
clusters = model.fit_predict(customers[cols_to_keep])
```

```python
def plot_elbow_curve(df, max_clusters=10):
    inertias = []
    
    for k in range(1, max_clusters+1):
        model = KMeans(n_clusters=k+1)
        cluster = model.fit_predict(df)
        inertias.append(model.inertia_)
        
    plt.figure(figsize=(12, 8))
    plt.plot(range(1, max_clusters+1), inertias, marker='o')
    plt.xticks(ticks=range(1, max_clusters+1), labels=range(1, max_clusters+1))
    plt.title('Inertia vs Number of Clusters')
    
    plt.tight_layout()
    plt.show()
    
    return inertias
  
inertias = plot_elbow_curve(customers[cols_to_keep])
print(inertias)
```

```python
def scaler(df):
    scaler = StandardScaler()
    scaler.fit(df)

    df_scaled = scaler.transform(df)
    return df_scaled

scaled_customers = scaler(customers[cols_to_keep])
inertias = plot_elbow_curve(scaled_customers)

print(inertias)
```

```python
customers_modif = customers.copy()
```

```python
customers_modif.drop('CustomerID', axis=1, inplace=True)
```

```python
customers_modif['Gender'] = customers_modif['Gender'].map({'Male': 1, 'Female': 0})
```

```python
scaled_customers = scaler(customers_modif)
```

```python
plot_elbow_curve(scaled_customers)
```

Choosing the best number of clusters is difficult from the graph as their was not any sharp decline to use

Here are some Possible solutions


* Gather more data.

* Look for new features.

* Look for business rules or insights that might lead to a decision for one cluster over another.

* Try different numbers of clusters, and see how the split goes in each of them.

The best possible numbers to chose are 4-6, we are going to use n_clusters=6 to go ahead

```python
model = KMeans(n_clusters=6)
```

```python
clusters = model.fit_predict(scaled_customers)
```

```python
customers['Cluster'] = clusters + 1
```

```python
print(customers['Cluster'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
print('m')


fig = plt.figure(figsize=(20, 10))
for i, column in enumerate(numeric_columns):
    df_plot = customers.groupby('Cluster')[column].mean()
    ax = fig.add_subplot(2, 2, i+1)
    ax.bar(df_plot.index, df_plot, color=sns.color_palette('Set1'), alpha=0.6)
    ax.set_title(f'Average {column.title()} per Cluster', alpha=0.5)
    ax.xaxis.grid(False)
    
plt.tight_layout()
plt.show()
```

From this graph we can see:
* Cluster one has an average age but are both high earners and spenders
* Cluster two are the oldest people by average but have relatively low annual income and spending score
* Cluster three has the lowest average age, lowest annual income but are averagely high spenders
* Cluster four has an averagely low age, average income but are quite large spenders
* Cluster five has an average age, High annual income but are the lowest spenders
* Cluster six has a high average age but an average annual income and spending score


Next we will plot the scatterplot of Annual Income over Age, Spending Score over Age and Spending Score over Annual Income. 

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 8))
sns.scatterplot(x='Age', y='Annual Income (k$)', hue='Cluster', data=customers, palette='Set1', alpha=0.4, ax=axs[0][0])

sns.scatterplot(x='Age', y='Spending Score (1-100)', hue='Cluster', data=customers, palette='Set1', alpha=0.4, ax=axs[0][1])

sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=customers, palette='Set1', alpha=0.4, ax=axs[1][0])

plt.tight_layout()
plt.show()
```

```python
plot_df = pd.crosstab(
  index=customers['Cluster'], columns=customers['Gender'],
  values=customers['Gender'], aggfunc='size', normalize='index'
)
```

```python
plot_df
```

```python
fig, ax = plt.subplots(figsize=(12,6))
plot_df.plot.bar(stacked=True, ax=ax, alpha=0.6)
ax.set_title(f'% Gender per Cluster', alpha=0.5)

ax.set_ylim(0, 1.4)
ax.legend(frameon=False)
ax.xaxis.grid(False)

plt.tight_layout()
plt.show()
```

From this chart we can see that 
* Clusters 3 and 6 have only female gender 
* Clusters 2, 4 and 5 have only male
* Cluster 5 has only a samll amount of males
